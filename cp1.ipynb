```python
import pandas as pd
import numpy as np
import csv
```

```python
def modify_s(t, y):
    gender=[]
    for i in t.iloc[:,6]:
        if i=='Male':
            gender.append(0)
        else:
            gender.append(1)
    t['Gender']=gender
    
    geography=[]
    for j in t.iloc[:,5]:
        if j=='S0':
            geography.append(0)
        elif j=='S1':
            geography.append(1)
        else:
            geography.append(2)
    t['Geography']=geography
    
    creditscore=[]
    credit_mean=t['CreditScore'].mean()
    credit_std=t['CreditScore'].std()
    for i in t.iloc[:,4]:
        creditscore.append((i-credit_mean)/credit_std)
    t['CreditScore']=creditscore
    
    balance=[]
    balance_mean=t['Balance'].mean()
    balance_std=t['Balance'].std()
    for i in t.iloc[:,9]:
        balance.append((i-balance_mean)/balance_std)
    t['Balance']=balance
    
    estimatedsalary=[]
    salary_mean=t['EstimatedSalary'].mean()
    salary_std=t['EstimatedSalary'].std()
    for i in t.iloc[:,13]:
        estimatedsalary.append((i-salary_mean)/salary_std)
    t['EstimatedSalary']=estimatedsalary
    
    age=[]
    for i in t.iloc[:,7]:
        age.append(i//y)
    t['Age']=age
    return t
```

```python
def modify_n(t, y):
    gender=[]
    for i in t.iloc[:,6]:
        if i=='Male':
            gender.append(0)
        else:
            gender.append(1)
    t['Gender']=gender
    
    geography=[]
    for j in t.iloc[:,5]:
        if j=='S0':
            geography.append(0)
        elif j=='S1':
            geography.append(1)
        else:
            geography.append(2)
    t['Geography']=geography
    
    creditscore=[]
    credit_min=t['CreditScore'].min()
    credit_max=t['CreditScore'].max()
    for i in t.iloc[:,4]:
        creditscore.append((i-credit_min)/(credit_max-credit_min))
    t['CreditScore']=creditscore
    
    balance=[]
    balance_min=t['Balance'].min()
    balance_max=t['Balance'].max()
    for i in t.iloc[:,9]:
        balance.append((i-balance_min)/(balance_max-balance_min))
    t['Balance']=balance
    
    estimatedsalary=[]
    salary_min=t['EstimatedSalary'].min()
    salary_max=t['EstimatedSalary'].max()
    for i in t.iloc[:,13]:
        estimatedsalary.append((i-salary_min)/(salary_max-salary_min))
    t['EstimatedSalary']=estimatedsalary
    
    age=[]
    for i in t.iloc[:,7]:
        age.append(i//y)
    t['Age']=age
    return t
```

```python
tr=pd.read_csv('train.csv')
te=pd.read_csv('test.csv')
#for KNN
tr_m=modify_n(tr, 4)
tr_1=tr_m.iloc[:,4:-1]
tr_2=tr_m.iloc[:,-1]
data=tr_1.values
labels=tr_2.values

te_m=modify_n(te, 4)
te=te_m.iloc[:,4:]
te_a=te.values
```

```python
tr=pd.read_csv('train.csv')
te=pd.read_csv('test.csv')
#for forest
tr_m5=modify_s(tr, 5)
tr_1=tr_m5.iloc[:,4:-1]
tr_2=tr_m5.iloc[:,-1]
data_5=tr_1.values
labels_5=tr_2.values

te_m=modify_s(te, 5)
te=te_m.iloc[:,4:]
te_a_5=te.values
```

```python
tr=pd.read_csv('train.csv')
te=pd.read_csv('test.csv')
#for MLP bayes
tr_m4=modify_s(tr, 4)
tr_1=tr_m4.iloc[:,4:-1]
tr_2=tr_m4.iloc[:,-1]
data_4=tr_1.values
labels_4=tr_2.values

te_m=modify_s(te, 4)
te=te_m.iloc[:,4:]
te_a_4=te.values
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(data)
data_scaled=scaler.transform(data)
te_a_scaled=scaler.transform(te_a)

knn=KNeighborsClassifier(n_neighbors=9, p=1, weights='distance')
knn.fit(data_scaled,labels)
ans_knn=knn.predict(te_a_scaled) #//4
```

```python
from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier(n_estimators=200, oob_score=True, n_jobs=-1, random_state=1, max_features='auto', min_samples_leaf=10)
forest.fit(data_5, labels_5)
ans_for=forest.predict(te_a_5) #//5
```

```python
from sklearn.neural_network import MLPClassifier
mlp=MLPClassifier(hidden_layer_sizes=(50,50), max_iter=200, alpha=1e-4, solver='sgd', verbose=10, tol=1e-4, random_state=1, learning_rate_init=0.1, momentum=0.5, learning_rate='adaptive')
mlp.fit(data_4, labels_4)
ans_mlp=mlp.predict(te_a_4) #//4
```

```python
from sklearn.naive_bayes import GaussianNB
clf=GaussianNB()
clf.fit(data_4, labels_4)
ans_bay=clf.predict(te_a_4) #//4
```
